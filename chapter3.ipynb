# Chapter 3: Adding Layers

In the previous chapter, we learned how to build a layer of neurons. Now we will extend our network by adding another layer, making it a two-layer neural network.

## 2-layer neural network

```python
inputs = [[1, 2, 3, 2.5], [2.0, 5.0, -1.0, 2], [-1.5, 2.7, 3.3, -0.8]]

weights = [[0.2, 0.8, -0.5, 1], [0.5, -0.91, 0.26, -0.5], [-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]

weights2 = [[0.1, -0.14, 0.5], [-0.5, 0.12, -0.33], [-0.44, 0.73, -0.13]]
biases2 = [-1, 2, -0.5]
```

```python
layer1_outputs = np.dot(inputs, np.array(weights).T) + biases

print(layer1_outputs)
```

```python
layer2_outputs = np.dot(layer1_outputs, np.array(weights2).T) + biases2

print(layer2_outputs)
```
